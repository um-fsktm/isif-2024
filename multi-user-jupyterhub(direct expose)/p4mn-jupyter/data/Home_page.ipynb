## 🧩 Training Overview

Welcome to the ​**5-Day P4 Programming Training** — a comprehensive, hands-on workshop designed to introduce you to the ​**P4 programming language** and the world of ​**software-defined, programmable networking**.

Over the course of this training, you will:

- Learn the fundamentals of the ​**P4 language** and how it enables ​**data plane programmability**.
- Develop and deploy ​**P4 programs** on emulated programmable switches (e.g., BMv2).
- Gain practical experience building network functions such as ​**IPv4 forwarding, tunneling, custom headers, source routing, and more**.
- Use tools like ​**Mininet, P4Runtime, and simple_switch** to simulate and test real-world network scenarios.

Whether you're a ​**student, researcher, network engineer, or developer**, this training will equip you with the skills to ​**design, implement, and innovate** at the data plane level using P4.
